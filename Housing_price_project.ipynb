```python
#IMPORTING NECESSARY LIBRARIES
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#READING THE CSV FILE
df=pd.read_csv("housing.csv")
print(df.head())
print(df.shape)
```

```python
print(df.info())
count=df["ocean_proximity"].value_counts()
print(count)
```

```python
#CONVERTING NOMINAL DATA INTO NUMERICAL DATA USING LABEL ENCODER
le=LabelEncoder()
df["ocean_proximity_encoded"]=le.fit_transform(df["ocean_proximity"])
count=df["ocean_proximity_encoded"].value_counts()
print(count)
```

```python
df.drop("ocean_proximity",axis=1,inplace=True)
```

```python
#CHECKING FOR MISSING OR NULL VALUES AND SUBSTITUTING IT WITH THE MEAN
has_null=df.isnull().any()
print(has_null)

df=df.fillna(df.mean())
has_null=df.isnull().any()
print(has_null)

```

```python
#SPLITTING THE DATA INTO DEPENDENT AND INDEPENDENT PARTS
x=df.drop("median_house_value",axis=1)
y=df["median_house_value"]
```

```python
#USING STANDARD SCALER TO THE SCALE THE DATA
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
print(x)
```

```python
#SPLITTING THE DATASET INTO TRAINING AND TEST SECTIONS
x_train,x_test,y_train, y_test=train_test_split(x,y, test_size=0.1, random_state=42 )

print(x_train.shape)
print(x_test.shape)
```

```python
#FITTING LINEAR REGRESSION MODEL
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#PREDICTING TRAIN DATA AND TEST DATA,  THEN SHOWING ACCURACY METRICS
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)

mse_train=mean_squared_error(y_train,y_train_pred,squared=False)
r2_train=r2_score(y_train,y_train_pred)
mse_test=mean_squared_error(y_test,y_test_pred,squared=False)
r2_test=r2_score(y_test,y_test_pred)
```

```python
print("MSE TRAIN:",mse_train)
print("R2 TRAIN:",r2_train)
print("MSE TEST:",mse_test)
print("R2 TEST:",r2_test)
```

```python
#ADDING POLYNOMIAL TERMS
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
```

```python
#SPLITTING THE DATA SET AFTER ADDING POLYNOMIAL FEATURES
x_train,x_test,y_train, y_test=train_test_split(x_poly,y, test_size=0.1, random_state=42 )
print(x_train.shape)
print(x_test.shape)
```

```python
#FITTING LINEAR REGRESSION MODEL
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#PREDICTING AND DISPLAYING ACCURACY METRICS FOR BOTH TRAIN AND TEST DATA
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)

mse_train=mean_squared_error(y_train,y_train_pred)
r2_train=r2_score(y_train,y_train_pred)
mse_test=mean_squared_error(y_test,y_test_pred)
r2_test=r2_score(y_test,y_test_pred)
print("MSE TRAIN:",mse_train)
print("R2 TRAIN:",r2_train)
print("MSE TEST:",mse_test)
print("R2 TEST:",r2_test)
```

```python
#USING STANDARD SCALER TO SCALE THE DATA AND PCA TU REDUCE THE DIMESNION
x=df.drop("median_house_value",axis=1)
y=df["median_house_value"]
scaler=StandardScaler()
x=scaler.fit_transform(x)
print(x)
from sklearn.decomposition import PCA
pca=PCA(n_components=8)
x_pca=pca.fit_transform(x)
print(x_pca.shape)

```

```python
#SPLITTING THE DATASET AND FITTING LINEAR REGRESSION MODEL
x_train,x_test,y_train, y_test=train_test_split(x_pca,y, test_size=0.1, random_state=42 )
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#PREDCITING AND SHOWING THE ACCUARAYC METRICES FOR BOTH TRAINING AND TESTING DATSET
y_train_pred=lr.predict(x_train)
y_test_pred=lr.predict(x_test)

mse_train=mean_squared_error(y_train,y_train_pred)
r2_train=r2_score(y_train,y_train_pred)
mse_test=mean_squared_error(y_test,y_test_pred)
r2_test=r2_score(y_test,y_test_pred)
print("MSE TRAIN:",mse_train)
print("R2 TRAIN:",r2_train)
print("MSE TEST:",mse_test)
print("R2 TEST:",r2_test)
```
